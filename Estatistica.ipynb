```python
import pandas as pd
import matplotlib.pyplot as plot
import numpy as np
```

```python
data = pd.read_csv("all_seasons.csv")
teste = data[(data['player_name'] == "Joel Embiid") & (data['season'].isin(["2022-23", '2021-2022', '2020-21', '2019-20'])) ]
print(teste)
```

```python

plot.plot(teste['season'], teste['pts'], marker='o')
plot.xlabel("Season")
plot.ylabel("Points per Game")
plot.title("Joel Embiid Points Per Game by Season")
plot.xticks(rotation=45, ha='right')
plot.show()
```

```python
temporada = data[(data['season'] == '2022-23')]
print(temporada)
```

```python
sorted_players = temporada.sort_values(by='reb', ascending=False).head(10)
players = sorted_players['player_name'].to_list()
rebounds = sorted_players['reb'].to_list()

plot.bar(players, rebounds)
plot.xlabel("Player")
plot.ylabel("Rebounds")
plot.title("Top 10 Players in Rebounds for the 2022-23 Season")
plot.xticks(rotation=45, ha='right')
plot.show()
```

```python
parametro1 = input("college: ")
parametro2 = input("team abbreviation: ")
if not (parametro1 in (data['college'].to_list())):
  print("college not found")
elif not (parametro2 in (data['team_abbreviation'].to_list())):
  print("team not found")
else:
  jogadores = data[(data['college'] == parametro1) & (data['team_abbreviation'] == parametro2)]
  jogadores = jogadores.sort_values(by = 'pts', ascending = False).head(10)
  nomes = jogadores['player_name'].to_list()
  pontos = jogadores['pts'].to_list()
  plot.bar(nomes,pontos)
  plot.xlabel("")
  plot.ylabel("pontos")
  plot.title("Maiores pontuadores de {} em {}".format(parametro1,parametro2))
  plot.xticks(rotation=45, ha='right')
  plot.show()
```
